```python
from collections import defaultdict
def dict_init():
    return defaultdict(list)
base_list = defaultdict(dict_init)
optionList = defaultdict(dict_init)

for fin in fin_type:
    for no in fin_no:
        max_page_no = float('inf')
        base_url = f"http://finlife.fss.or.kr/finlifeapi/{fin}.json?auth={FSS_API_KEY}&topFinGrpNo={no}&pageNo="
        pageNo = 0

        while max_page_no > pageNo:
            pageNo += 1
            response = requests.get(base_url + str(pageNo))
            if response.status_code != 200:
                break
            response = response.json()['result']
            for i in range(len(response['baseList'])):
                response['baseList'][i]['is_diposit'] = fin_type[fin]
            max_page_no = min(max_page_no, response['max_page_no'])
            print(max_page_no)
            if not response['baseList']:
                continue
            base_list[fin][no] += response['baseList']
            optionList[fin][no] += response['optionList']
```
